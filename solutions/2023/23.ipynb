```python
def find_trails(hiking, node, end, path=[]):
    nrows, ncols = hiking.shape
    cr, cc = node
    path = path + [(cr, cc)]
    if (cr, cc) == end:
        return [path]
    if not (0 <= cr < nrows and 0 <= cc < ncols) or hiking[cr, cc] == "#":
        return []
    
    paths = []
    neighbours = []
    moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    for mr, mc in moves:
        nr, nc = cr + mr, cc + mc
        if hiking[nr, nc] == ".":
            neighbours.append((nr, nc))
        elif (mr, mc) == (0, 1) and hiking[nr, nc] == ">":
            neighbours.append((nr, nc + 1))
        elif (mr, mc) == (1, 0) and hiking[nr, nc] == "v":
            neighbours.append((nr + 1, nc))

    for nr, nc in neighbours:
        if (nr, nc) not in path:
            next_paths = find_trails(hiking, (nr, nc), end, path)
            for next_path in next_paths:
                paths.append(next_path)
    return paths       


paths = find_trails(hiking, start, end, path=[])

def trace_path_nodes(paths):
    path_lengths = {k: 0 for k in range(len(paths))}
    for k, path in enumerate(paths):
        for i in range(len(path) - 1):
            cr, cc = path[i]
            nr, nc = path[i + 1]
            path_lengths[k] += abs(nr - cr) + abs(nc - cc)
    return path_lengths

max(trace_path_nodes(paths).values())
```

```python
def find_corners_junctions(hiking):
    nrows, ncols = hiking.shape
    corners = []
    junctions = []
    for r in range(1, nrows - 1):
        for c in range(1, ncols - 1):
            # import pdb; pdb.set_trace()
            if hiking[r, c] == "." and not (
                np.array_equal(hiking[r, c - 1:c + 2], [".", ".", "."])
                or np.array_equal(hiking[r - 1:r + 2, c], [".", ".", "."])
            ):
                corners.append((r, c))
            elif ((
                    np.array_equal(hiking[r, c - 1:c + 2], [".", ".", "."])
                    and not np.array_equal(hiking[r - 1:r + 2, c], ["#", ".", "#"])
                )
                or (
                    np.array_equal(hiking[r - 1:r + 2, c], [".", ".", "."])
                    and not np.array_equal(hiking[r, c - 1:c + 2], ["#", ".", "#"])
                )
            ):
                junctions.append((r, c))
    return corners, junctions


def organize_corners_junctions(corners, junctions):
    by_row = defaultdict(list)
    by_col = defaultdict(list)
    for r, c in corners + junctions:
        by_row[r].append((r, c))
        by_col[c].append((r, c))
    return by_row, by_col


corners, junctions = find_corners_junctions(hiking)
by_row, by_col = organize_corners_junctions(corners, junctions)

display(hiking)
```

```python
def turns_only_graph(hiking, corners, junctions, start_edge, end_edge):
    hiking_turns = hiking.copy()
    for r, c in corners + junctions:
        hiking_turns[r, c] = "c"
    
    H = nx.Graph()
    H.add_edge(*start_edge, length=1)
    H.add_edge(*end_edge, length=1)
    H.add_node(start_edge[0], junction=True) # not true junctions
    H.add_node(end_edge[1], junction=True)
    corners = set(corners)
    junctions = set(junctions)
    for cr, cc in corners | junctions:
        check_coords = by_row[cr] + by_col[cc]
        for nr, nc in check_coords:
            if (cr, cc) == (nr, nc):
                continue
            dr = nr - cr
            dc = nc - cc
            if dr > 0:
                check = hiking_turns[cr+1:nr, cc]
            elif dr < 0:
                check = hiking_turns[nr+1:cr, cc]
            elif dc > 0:
                check = hiking_turns[cr, cc+1:nc]
            elif dc < 0:
                check = hiking_turns[cr, nc+1:cc]
            
            if all(ch == "." for ch in check):
                H.add_edge((cr, cc), (nr, nc), length=abs(nr - cr) + abs(nc - cc))
                H.add_node((cr, cc), junction=(cr, cc) in junctions)
                H.add_node((nr, nc), junction=(nr, nc) in junctions)
    return H


H = turns_only_graph(hiking, corners, junctions, (start, (1, 1)), ((nrows - 2, ncols - 2), end))
```

```python
# should be doing graph algorithm, very brute force
S = nx.Graph()
for j1 in junctions + [start, end]:
    for j2 in junctions + [start, end]:
        if j1 == j2:
            continue
        junction_path = nx.shortest_path(H, j1, j2)
        check = [H.nodes[node]["junction"] for node in junction_path][1:-1]
        if not any(check):
            length = 0
            for n in range(len(junction_path) - 1):
                l1 = junction_path[n]
                l2 = junction_path[n + 1]
                length += H.get_edge_data(l1, l2)["length"]
            S.add_edge(j1, j2, length=length)    
```
